```python
import seaborn as sns
import matplotlib.pyplot as plt

def boxplt(data, column):
    """Generate a boxplot for the specified column in the DataFrame."""
    sns.boxplot(data=data[column])
    plt.title(f'Boxplot of {column}')
    plt.ylabel(column)
    plt.show()

boxplt(data, "Potential_overall")

```
